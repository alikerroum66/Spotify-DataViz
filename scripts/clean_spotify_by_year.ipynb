Step 3:Clean numeric columns properly

```python
df['Spotify_Revenue_EUR_Billion'] = (
    df['Spotify_Revenue_EUR_Billion']
    .astype(str)
    .str.replace(r"[^\d.]", "", regex=True)
    .astype(float)
)

df['Song_Streams_Estimate'] = pd.to_numeric(
    df['Song_Streams_Estimate'],
    errors='coerce'
)

df['Top_Artist_Yearly_Streams_Billion'] = pd.to_numeric(
    df['Top_Artist_Yearly_Streams_Billion'],
    errors='coerce'
)

df['Year'] = df['Year'].astype(int) 

```

Step 4:Clean text columns

```python
text_cols = ['Top_Artist', 'Most_Popular_Song']

for col in text_cols:
    df[col] = df[col].str.strip()

df.info()
```
